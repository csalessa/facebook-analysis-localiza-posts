```python
#retirando acento
def tira_acento(texto):
    texto = texto.replace("Ã¡", "a").replace("Ã£", "a").replace("Ã ", "a").replace("Ã©", "e").replace("Ãª", "e").replace("Ã­", "i").replace("Ã´", "o").replace("Ã³", "o").replace("Ãµ", "o").replace("Ãº", "u").replace("Ã»", "u").replace("Ã§", "c")
    return texto
fb["text"]=fb["text"].apply(tira_acento)
fb
```

```python
#retirar emoji
import emoji as emj
EMOJIS = emj.UNICODE_EMOJI["en"]

def extract_emoji(fb):
    for index, row in fb.iterrows():
        for emoji in EMOJIS:
            if emoji in row["text"]:
                row["text"] = row["text"].replace(emoji, " ")
              

    extract_emoji(fb)
fb
```
